```python
import pandas as pd
import numpy as np
```

```python
data = {
    'StudentID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math_Score': [30, 20, None, 45, None],
    'English_Score': [40, None, 35, 30, 43]
}
df = pd.DataFrame(data)
df
```

```python
df.head(3)
```

```python
df.tail(2)
```

```python
print(df.count())
```

```python
print("Sum of Math Scores:",df['Math_Score'].sum())
print("Mean of Math Scores:",df['Math_Score'].mean())
print("Median of Math Scores:",df['Math_Score'].median())
```

```python
dt={
    'A':[10,-20,30,-40,30,10,10]
}
df1=pd.DataFrame(dt)
mod=df1['A'].mode()
print("Mode=",mod)
```

```python
print("Standard Deviation of Math Scores:",df['Math_Score'].std())
```

```python
print("Minimum value in A:",df1['A'].min())
```

```python
print("Maximum value in A:",df1['A'].max())
```

```python
abs=df1.abs()
print("before:\n",df1)
print("After:\n",abs)
```

```python
print("Product of values:",df1['A'].prod())
```

```python
print("Cumulative sum of values:\n",df1['A'].cumsum())
```

```python
print("Cumulative product of values:\n",df1.cumprod())
```
