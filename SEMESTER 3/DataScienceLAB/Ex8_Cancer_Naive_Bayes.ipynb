```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn import metrics
```

```python
cancer=pd.read_csv("Cancer_DS.csv")
```

```python
cancer.head()
```

```python
cancer.drop(["id","Unnamed: 32"],axis=1,inplace=True)
```

```python
cancer.head()
```

```python
feature=cancer[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
labels=cancer[['diagnosis']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(feature,labels,test_size=0.2,random_state=42)
```

```python
NB=GaussianNB()
NB.fit(X_train,y_train)
```

```python
Y_pred=NB.predict(X_test)
```

```python
print(Y_pred)
```

```python
accuracy=metrics.accuracy_score(y_test,Y_pred)
```

```python
print(f"Accuracy:{accuracy:.2f}")
```
