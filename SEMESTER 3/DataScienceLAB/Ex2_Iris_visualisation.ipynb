```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.describe()
```

```python
iris['sepal.length'].hist(color="red")
plt.show()
```

```python
iris['sepal.length'].plot(kind='box')
plt.show()
```

```python
# Scatter plot of sepal length vs. sepal width
plt.scatter(iris['sepal.length'], iris['petal.length'],color="pink")
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter Plot of Sepal Length vs. Petal Length')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris, figsize=(10, 10))
plt.show()
```

```python
plt.scatter(iris['sepal.length'], iris['sepal.width'], s=50*iris['petal.length'], color="violet",alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Bubble Chart of Sepal Length vs. Sepal Width (Petal Length as Bubble Size)')
plt.show()
```

```python

# Create a 2D density chart for sepal length and sepal width
sns.kdeplot(data=iris, x='sepal.length', y='sepal.width', cmap='Blues', shade=True)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('2D Density Chart of Sepal Length and Sepal Width')
plt.show()


```

```python
# Parallel coordinates plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(iris, 'variety', colormap='viridis')
plt.title('Parallel Coordinates Plot of Iris Dataset')
plt.show()
```

```python
reference_value = iris['sepal.length'].mean()

# Calculate deviations
iris['deviation'] = iris['sepal.length'] - reference_value

# Create a deviation bar chart
plt.figure(figsize=(10, 6))
plt.bar(iris['variety'], iris['deviation'], color='skyblue')
plt.xlabel('Species')
plt.ylabel('Deviation from Mean Sepal Length')
plt.title('Deviation Chart of Sepal Length from Mean')
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(iris, 'variety', colormap='viridis')
plt.title('Andrews Curves for Iris Dataset')
plt.show()
```

```python
sns.set(style="whitegrid")

# Create a distribution chart for sepal length
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x="sepal.length", kde=True, color='skyblue')
plt.xlabel('Sepal Length')
plt.title('Distribution Chart of Sepal Length')
plt.show()
```
